# Research Question

How do track characteristics, driver performance metrics, and weather conditions affect the probability of winning a Formula 1 race?


# Background and Prior Work

## Background & Introduction to Formula 1 Racing
Formula 1 racing combines high-speed competition with complex engineering, where predictive analytics play a crucial role in decision-making. Teams utilize vast amounts of data, from tire degradation to driver performance, to optimize race strategies and improve outcomes.

## Prior Work
1. **Application of Machine Learning**: Recent studies, such as the one documented by Aalto University, have explored the use of machine learning algorithms like Support Vector Machines, Random Forest, and Neural Networks to predict the timing of pit stops in F1 races. This research highlights the importance of various race factors, including tire degradation and weather conditions, in decision-making processes. ([Aaltodoc](https://aaltodoc.aalto.fi/server/api/core/bitstreams/70d5a580-c282-4278-8462-94d061471546/content))

2. **Comprehensive Data Analysis Platforms**: Platforms like the one found in the GitHub repository 'f1-analysis' use the FastF1 library to manipulate telemetry and race data for in-depth insights into race performance and strategies. This repository serves as a valuable resource for enthusiasts and analysts looking to perform statistical analyses and create data visualizations based on F1 race data. ([f1-analysis](https://github.com/lalutir/f1-analysis))

3. **Industrial Impact of Predictive Analytics**: An overview provided by sites like Penn State's page on predictive analytics in F1 discusses the industrial relevance and financial impact of predictive analytics in the sport. It outlines how teams leverage real-time data from over 300 sensors on each car to make strategic decisions that can alter race results, illustrating the blend of data science with competitive sports to enhance team performance and race strategy. ([F1 and predictive analysis](https://sites.psu.edu/aboutoliviadisanti/2023/11/14/unleashing-the-power-of-data-science-in-formula-1/))


# Hypothesis


We predict that track characteristics, driver performance metrics, and weather conditions will each have a significant influence on the probability of winning a Formula 1 race. These factors are hypothesized to interact, such that their combined impact is more predictive of race outcomes than any single factor alone.

This hypothesis is based on the understanding that Formula 1 racing is a complex sport where various elements such as the physical layout of the track, the skills and current form of the driver, and unpredictable elements like weather dramatically affect performance. Previous studies and data analyses, such as those referenced in existing research on machine learning applications in F1 (e.g., the study at Aalto University), suggest that these variables not only affect race strategy but also directly influence race results through their impact on vehicle handling, tire wear, and driver safety. The interaction of these variables is likely complex, given the dynamic and highly conditional nature of racing, where the optimal strategy must continuously adapt to the evolving race conditions. ([Aaltodoc](https://aaltodoc.aalto.fi/server/api/core/bitstreams/70d5a580-c282-4278-8462-94d061471546/content))


# Data

[Formula 1 World Championship (1950 - 2024)](https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020?select=driver_standings.csv):

This dataset consists of the information on Formula 1 races, drivers, lap times, constructors, pit stops from the championship in 1950 until the latest 2024 season. This data allows us to be able to answer our research question because we can see what contributes to whether someone will win the race with the data of lap times and how the information of the drivers can help us see if there are any attributes that allows people to predict the winner of a race.

[Formula 1 World Championship History (1950-2024)](https://www.kaggle.com/datasets/muhammadehsan02/formula-1-world-championship-history-1950-2024):

This is a comprehensive dataset that is organized with detail in the performance of constructors and drivers. As well as, showcasing who topped the charts and the results of the races. This would be useful because we are able to make inferences about those who were able to top the charts and how their performances coincide with their ability to make it to the leaderboard. 

[Formula 1 Dataset (1950-2023)](https://www.kaggle.com/datasets/suletanmay/formula-1-dataset-1950-2023-cleaned): 

This particular dataset looks at the history and evolution of the sport from the achievements of renowned drivers and teams. Being able to access these statistics, can let us see what the factors are that contribute to predicting a Formula 1 driver from winning by seeing the improvement throughout history on how drivers are able to win.
