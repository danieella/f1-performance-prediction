# Hypothesis


Our hypothesis is that track characteristics, driver performance metrics and pit stop times each significantly influence the probability of winning a Formula 1 race. 

Track features, such as the layout and corner types, can favor different car setups, potentially creating positive or negative correlations with winning likelihood depending on the car’s strengths. For instance, high-speed tracks may positively correlate with cars optimized for straight-line speed, while technical circuits could benefit those with strong downforce. Driver metrics, including qualifying position, overtaking ability, and consistency, are expected to correlate positively with success, as they reflect a driver’s ability to maximize their position and minimize errors under varying race conditions. Weather conditions add a layer of unpredictability, where rain, temperature, and humidity can shift race dynamics. Rain, for example, may benefit drivers skilled in wet conditions, while high temperatures could challenge teams with poor tire or cooling management, introducing both positive and negative correlations. 

Previous studies and data analyses, such as those referenced in existing research on machine learning applications in F1 (e.g., the study at Aalto University), suggest that these variables not only affect race strategy but also directly influence race results through their impact on vehicle handling, tire wear, and driver safety. The interaction of these variables is likely complex, given the dynamic and highly conditional nature of racing, where the optimal strategy must continuously adapt to the evolving race conditions.([Aaltodoc](https://aaltodoc.aalto.fi/server/api/core/bitstreams/70d5a580-c282-4278-8462-94d061471546/content))

## Analyzing Driver Performance Metrics and Their Impact on Race Outcomes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
driver_standings_df = pd.read_csv('cleaned_driver_standings.csv')

# Distribution of Points, Wins, and Positions
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

driver_standings_df['points'].hist(bins=50, ax=axes[0])
axes[0].set_title('Distribution of Points')
axes[0].set_xlabel('Points')
axes[0].set_ylabel('Frequency')

driver_standings_df['wins'].hist(bins=20, ax=axes[1])
axes[1].set_title('Distribution of Wins')
axes[1].set_xlabel('Wins')
axes[1].set_ylabel('Frequency')

driver_standings_df['position'].hist(bins=20, ax=axes[2])
axes[2].set_title('Distribution of Positions')
axes[2].set_xlabel('Position')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap for Performance Metrics
corr = driver_standings_df[['points', 'position', 'wins']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Driver Metrics')
plt.show()
```

```python
results_df = pd.read_csv('cleaned_results.csv')
```

```python
# Grid Position vs Final Position
plt.figure(figsize=(10, 6))
sns.scatterplot(data=results_df, x='grid_position', y='position', alpha=0.5)
plt.title('Grid Position vs Final Position')
plt.xlabel('Starting Grid Position')
plt.ylabel('Final Position')
plt.grid(True)
plt.show()
```

Interpretation:
A weak positive trend is observed: drivers starting closer to the front (lower grid positions) generally achieve better results (lower final positions).
Outliers indicate drivers who improved significantly or underperformed relative to their starting grid.

```python
# Distribution of Points for Drivers by Final Position
plt.figure(figsize=(10, 6))
sns.boxplot(data=results_df, x='position', y='points')
plt.title('Distribution of Points by Final Position')
plt.xlabel('Final Position')
plt.ylabel('Points')
plt.grid(True)
plt.show()
```

Interpretation: Top Positions (1st to 3rd): Drivers finishing in the top positions consistently earn the most points, with 1st place offering significantly higher rewards.
Points Decline: Points drop steeply after the top 5 and taper off; positions outside the top 10 rarely score points.
Variability: Top positions show variability in points due to differing race outcomes, while lower positions see little to no scoring.

```python
# Relationship Between Grid Position and Wins
winning_drivers = results_df[results_df['position'] == 1]
plt.figure(figsize=(10, 6))
sns.histplot(winning_drivers['grid_position'], bins=20, kde=True)
plt.title('Starting Grid Position of Winning Drivers')
plt.xlabel('Grid Position')
plt.ylabel('Frequency')
plt.show()
```

Most race winners started in the top grid positions, particularly in the first 5 positions.
The distribution has a long tail, showing that some drivers win from midfield or backfield positions, though this is less common.

```python
# Aggregate Driver Performance: Total Points vs Wins
driver_performance = driver_standings_df.groupby('driverId').agg({
    'points': 'sum',
    'wins': 'sum',
    'position': 'mean'
}).reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=driver_performance, x='points', y='wins', alpha=0.7)
plt.title('Total Points vs Total Wins for Drivers')
plt.xlabel('Total Points')
plt.ylabel('Total Wins')
plt.grid(True)
plt.show()
```

Driver consistency across multiple races is an essential indicator of success in Formula 1. This analysis focuses on how cumulative points and wins correlate with average standings position, providing insights into how consistent performances contribute to long-term success.

```python
# Aggregate Driver Performance: Total Points vs Wins
driver_performance = driver_standings_df.groupby('driverId').agg({
    'points': 'sum',
    'wins': 'sum',
    'position': 'mean'
}).reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=driver_performance, x='points', y='wins', alpha=0.7)
plt.title('Total Points vs Total Wins for Drivers')
plt.xlabel('Total Points')
plt.ylabel('Total Wins')
plt.grid(True)
plt.show()
```

Interpretation:
A clear positive correlation exists between total points and total wins.
Drivers with more wins typically accumulate more points, reflecting the scoring system's emphasis on winning races.

```python
# Average Position vs Wins
plt.figure(figsize=(10, 6))
sns.lineplot(data=driver_performance, x='wins', y='position', marker='o')
plt.title('Average Position vs Total Wins')
plt.xlabel('Total Wins')
plt.ylabel('Average Position (Lower is Better)')
plt.grid(True)
plt.show()
```

Interpretation:
Drivers with more wins tend to have lower average positions, showcasing their ability to consistently finish in top positions.
The relationship flattens for drivers with very high win counts, suggesting diminishing returns in further improving their average standings.

```python
# Points Distribution for Drivers with Low Average Positions
consistent_drivers = driver_performance[driver_performance['position'] <= 3]  # Drivers with average positions <= 3
plt.figure(figsize=(10, 6))
sns.boxplot(data=consistent_drivers, x='wins', y='points')
plt.title('Points Distribution for Drivers with High Consistency')
plt.xlabel('Total Wins')
plt.ylabel('Total Points')
plt.grid(True)
plt.show()
```

Interpretation:
Drivers with high consistency (low average position) show a strong overlap in points distribution, regardless of win count.
Consistency in scoring points across races is a hallmark of top-performing drivers, even if they don't always secure wins.

## Analyzing Pit Stop Times and Their Impact on Race Outcomes

I am trying to create visualizations to explore the relationships between pit stop times and race outcomes in Formula 1. I believe pit stop efficiency influences race strategies and performance and by analyzing the distribution of pit stop durations, their correlation with final race positions and points, and variations across circuits and years, we might be able to prove it.

```python
pit_stops = pd.read_csv('cleaned_pit_stops.csv')
results = pd.read_csv('cleaned_results.csv')
races = pd.read_csv('cleaned_races.csv')
drivers = pd.read_csv('cleaned_drivers.csv')
circuits = pd.read_csv('cleaned_circuits.csv')
driver_standings = pd.read_csv('cleaned_driver_standings.csv')

pit_stop_results = pit_stops.merge(results, on=['raceId', 'driverId'], how='inner')

pit_stop_results_races = pit_stop_results.merge(races, on='raceId', how='inner')

full_data = pit_stop_results_races.merge(drivers, on='driverId', how='inner')
```

```python
# Distribution of Pit Stop Times
plt.figure(figsize=(10, 6))
sns.histplot(pit_stops['milliseconds'], bins=30, kde=True)
plt.title("Distribution of Pit Stop Times (in milliseconds)")
plt.ylabel("Pit Stop Duration (ms)")
plt.xlabel("Frequency")
plt.show()
```

The graph illustrates the distribution of pit stop times in milliseconds. This graph connects to the research question as it provides an overview of how pit stop durations vary, helping identify typical and outlier times. Such insights are crucial to understanding the role of pit stop efficiency in influencing race outcomes and whether extremely long or short stops correlate with a driver's ability to secure a win.

```python
# Average Pit Stop Time by Final Race Position
avg_pit_stop_by_position = full_data.groupby('position')['milliseconds'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_pit_stop_by_position, x='position', y='milliseconds')
plt.title("Average Pit Stop Time by Final Race Position")
plt.xlabel("Final Position")
plt.ylabel("Average Pit Stop Time (ms)")
plt.xticks(rotation=90)
plt.show()
```

This graph shows the average pit stop time for each final race position, directly addressing the research question by revealing the relationship between pit stop efficiency and race outcomes. Faster pit stops generally correlate with better race positions, suggesting their significance in achieving competitive results. It emphasizes how critical pit stop performance is for increasing the probability of winning or finishing in higher positions. There are not many finsihes at positions after 18 and hence, the avergae pit stop time might be going down. 

```python
# Correlation Between Pit Stop Times and Race Points
plt.figure(figsize=(10, 6))
sns.scatterplot(data=full_data, x='milliseconds', y='points')
plt.title("Correlation Between Pit Stop Times and Race Points")
plt.xlabel("Pit Stop Duration (ms)")
plt.ylabel("Race Points")
plt.show()
```

This scatter plot explores the relationship between pit stop durations and race points. It directly addresses the research question by examining whether longer or shorter pit stops correlate with higher race points, a critical factor in determining the probability of winning. While the spread indicates variability, clusters may reveal trends, such as consistently short pit stops among top-scoring drivers. This graph is highly relevant for analyzing the impact of pit stop efficiency on race performance and outcomes.

```python
# Average Pit Stop Times by Circuit
avg_pit_stop_by_circuit = full_data.groupby('circuitId')['milliseconds'].mean().reset_index()
avg_pit_stop_by_circuit = avg_pit_stop_by_circuit.merge(circuits, on='circuitId', how='inner')

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_pit_stop_by_circuit, x='circuit_name', y='milliseconds')
plt.title("Average Pit Stop Times by Circuit")
plt.xlabel("Circuit Name")
plt.ylabel("Average Pit Stop Time (ms)")
plt.xticks(rotation=90)
plt.show()
```

This bar chart depicts the average pit stop times for each circuit, connecting directly to the research question by exploring how track characteristics influence pit stop efficiency. Certain circuits may have consistently longer or shorter average pit stop times due to their layout, pit lane design, or weather conditions. This graph is highly relevant for understanding how track-specific factors affect pit stops, which in turn impacts race strategy and the probability of winning.

The insights reveal the critical role of pit stops in determining a driver's success, with shorter times often associated with better outcomes, while also showcasing the impact of track-specific features and temporal trends. Together, these graphs should help us building a model because it help us understand how pit stops interact with broader race dynamics to shape winning probabilities in Formula 1.
